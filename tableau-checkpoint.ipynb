```python
# Install the necessary SQL library
!pip install sqlalchemy sqlite3 pandas

# Import necessary libraries
import sqlite3
import pandas as pd

# Create a SQLite database (simulating a real SQL database)
conn = sqlite3.connect('sales_data.db')
cursor = conn.cursor()

# Create Customers, Products, and Purchases tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Customers (
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT,
    ContactInfo TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Products (
    ProductID INTEGER PRIMARY KEY,
    ProductName TEXT,
    ProductCategory TEXT,
    Price REAL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Purchases (
    PurchaseID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    ProductID INTEGER,
    Quantity INTEGER,
    PurchaseDate TEXT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID),
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
)
''')

# Simulating sample data
customers_data = [
    (1, 'Alice', 'alice@example.com'),
    (2, 'Bob', 'bob@example.com'),
    (3, 'Charlie', 'charlie@example.com')
]

products_data = [
    (1, 'Laptop', 'Electronics', 1000),
    (2, 'Smartphone', 'Electronics', 700),
    (3, 'Tablet', 'Electronics', 500)
]

purchases_data = [
    (1, 1, 1, 1, '2024-01-15'),
    (2, 2, 2, 2, '2024-01-16'),
    (3, 3, 3, 1, '2024-01-17'),
    (4, 1, 2, 3, '2024-01-18')
]

# Insert data into the tables
cursor.executemany('INSERT INTO Customers VALUES (?,?,?)', customers_data)
cursor.executemany('INSERT INTO Products VALUES (?,?,?,?)', products_data)
cursor.executemany('INSERT INTO Purchases VALUES (?,?,?,?,?)', purchases_data)

# Commit and close connection
conn.commit()
```

```python
# Query the normalized data
query = '''
SELECT Customers.CustomerName, Products.ProductName, SUM(Purchases.Quantity) AS TotalQuantity,
       SUM(Purchases.Quantity * Products.Price) AS TotalSpend
FROM Purchases
JOIN Customers ON Purchases.CustomerID = Customers.CustomerID
JOIN Products ON Purchases.ProductID = Products.ProductID
GROUP BY Customers.CustomerName, Products.ProductName
ORDER BY TotalSpend DESC
'''

# Load the data into a pandas DataFrame
df = pd.read_sql(query, conn)

# Display the normalized data
df.head()
```

```python
# Install required visualization libraries
!pip install seaborn matplotlib

# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for total spend per customer and product
plt.figure(figsize=(10,6))
sns.barplot(x='TotalSpend', y='CustomerName', data=df, hue='ProductName', palette='Blues_d')
plt.title('Total Spend per Customer by Product')
plt.xlabel('Total Spend ($)')
plt.ylabel('Customer Name')
plt.legend(title='Product')
plt.show()

# Pie chart for the share of total spend per product
product_share = df.groupby('ProductName')['TotalSpend'].sum().reset_index()

plt.figure(figsize=(7,7))
plt.pie(product_share['TotalSpend'], labels=product_share['ProductName'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Product Spend Distribution')
plt.show()
```

```python
# Export the data to CSV for Tableau import
df.to_csv('sales_data_for_tableau.csv', index=False)
```
